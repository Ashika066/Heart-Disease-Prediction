<b> Assigning inputs and Outputs <b>

```python
X=Heart_disease[['glucose','diabetes','diaBP','prevalentHyp','sysBP','age']].values
Y=Heart_disease[['TenYearCHD']].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train
```

```python
Y_train
```

```python
model=LogisticRegression(C=100,penalty='l2',solver='liblinear')
model.fit(X_train,Y_train)
```

<b> Predictions <b>

```python
model.predict(X_test[0:10])
```

```python
Predictions_made=model.predict(X_test)
```

```python
Predictions_made
```

```python
model.score(X_train,Y_train)
```

<b><u> Reporting the quality value and accuracy of model <b>

```python
Score_of_Training_Data=model.score(X_train,Y_train)
print(Score_of_Training_Data)
```

```python
Score_of_Testing_Data=model.score(X_test,Y_test)
print(Score_of_Testing_Data)
```

```python
print('Training Accuracy:%.2f'% model.score(X_train,Y_train))
print('Test Accuracy:%.7f'% model.score(X_test,Y_test))
```

<b>Confusion Matrix


```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
Confusion_Matrix=metrics.confusion_matrix(Y_test,Predictions_made)
print(Confusion_Matrix)
```

<b>Using seaborn to create confusion matrix

```python
plt.figure(figsize=(9,9))
sns.heatmap(Confusion_Matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Purples');
plt.ylabel('Actual label(Have risk of TenYearCHD/Not have risk of TenYearCHD)');
plt.xlabel('Predicted label(Have risk of TenYearCHD/Not have risk of TenYearCHD)');
all_sample_title = 'Accuracy Score: {0}'.format(Score_of_Testing_Data)
plt.title(all_sample_title,size=15);
```
